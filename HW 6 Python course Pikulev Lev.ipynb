Задание 1
Используем файл keywords.csv.

Необходимо написать гео-классификатор, 
который каждой строке сможет выставить географическую принадлежность 
определенному региону. Т. е. если поисковый запрос содержит название 
города региона, то в столбце 'region' пишется название этого региона. 
Если поисковый запрос не содержит названия города, 
то ставим 'undefined'.

Правила распределения по регионам Центр, Северо-Запад и Дальний Восток:

geo_data = {

    'Центр': ['москва', 'тула', 'ярославль'],

    'Северо-Запад': ['петербург', 'псков', 'мурманск'],

    'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']

}

Результат классификации запишите в отдельный столбец region.

```python
import pandas as pd
```

```python
#посмотрим на датафрейм
keywords = pd.read_csv('keywords.csv')
keywords.head()
```

```python
#разобъем текст запроса на списки по разделителю ' '
# и попробуем написать заготовку для функции
for row in keywords['keyword'].apply(lambda x: x.split(' ')):
    if 'москва'.lower() in row or 'тула'.lower() in row or 'ярославль'.lower() in row:
        print('Центр')
    elif 'петербург'.lower() in row or 'псков'.lower() in row or 'мурманск'.lower() in row:
        print('Северо-Запад')
    elif 'владивосток'.lower() in row or 'сахалин'.lower() in row or 'хабаровск'.lower() in row:
        print('Дальний Восток')
    else:
        print('undefined')
```

```python
def geo_class(row):
    """функция принимает строку и по полю 'keyword' выдает регион в 
    зависимости от города, название которого содержится в запросе
    """
    row = row['keyword'].split(' ')
    if 'москва'.lower() in row or 'тула'.lower() in row or 'ярославль'.lower() in row:
        return 'Центр'
    elif 'петербург'.lower() in row or 'псков'.lower() in row or 'мурманск'.lower() in row:
        return 'Северо-Запад'
    elif 'владивосток'.lower() in row or 'сахалин'.lower() in row or 'хабаровск'.lower() in row:
        return 'Дальний Восток'
    else:
        return 'undefined'
```

```python
#запишем результат в поле 'region'
keywords['region'] = keywords.apply(geo_class, axis=1)
#Проверим для регииона 'Дальний Восток'
keywords[keywords.region == 'Дальний Восток'].head(10)
```

```python
# сделаем небольшую проверку, сходится ли количество записей о регионе с 
# количеством записей всего (все ли подписаны)
len(keywords[keywords.region == 'Центр']) + \
    len(keywords[keywords.region == 'Северо-Запад']) + \
        len(keywords[keywords.region == 'Дальний Восток']) + \
            len(keywords[keywords.region == 'undefined']) == len(keywords)
```

Задание 2
Напишите функцию, которая классифицирует фильмы из материалов занятия по следующим правилам:
    - оценка 2 и меньше - низкий рейтинг
    - оценка 4 и меньше - средний рейтинг
    - оценка 4.5 и 5 - высокий рейтинг

Результат классификации запишите в столбец class

```python
ratings = pd.read_csv('ml-latest-small/ratings.csv')
ratings.head()
```

```python
ratings[['userId', 'rating']].head()
```

```python
def classificator(row):
    row = row['rating']
    if row <= 2:
        return('низкий рейтинг')
    if row <= 4:
        return('средний рейтинг')
    else:
        return('высокий рейтинг')
```

```python
#запишем результат в столбец 'class'
ratings['class'] = ratings.apply(classificator, axis=1)
ratings.head(10)
```

```python
#Проверим для высоких оценок 'высокий рейтинг'
ratings[ratings['class'] == 'высокий рейтинг'].head(15)
```

Задание 3
Посчитайте среднее значение Lifetime киноманов 
(пользователи, которые поставили 100 и более рейтингов). 
Под Lifetime понимается разница между максимальным и минимальным 
значением timestamp для каждого пользователя. Ответ дайте в днях.

```python
#нужен датафрейм из двух полей 'userId', 'rating', сгруппированный по userId
# считаем кол-во строк и сбрасываем индекс
ratings[['userId', 'rating']].groupby(['userId']).count().reset_index().head()
```

```python
# приравниваем это к переменной ratings_filtered
ratings_filtered = ratings[['userId', 'rating']].groupby(['userId']).count().reset_index()

ratings_filtered.head()
```

```python
# выставляем фильтр, чтобы получить только тех пользователей, кто поставил 100 и более оценок
ratings_filtered[(ratings_filtered['rating'] >= 100)].head()
```

```python
# выводим один столбец userId и при помощи tolist() создаем список
ratings_filtered[(ratings_filtered['rating'] >= 100)]['userId'].tolist()
```

```python
# обозначаем этот список как active_users
active_users = ratings_filtered[(ratings_filtered['rating'] >= 100)]['userId'].tolist()
```

```python
# вот датафрейм, в котором будем считать максимальный и минимальный timestamp
ratings[ratings['userId'].isin(active_users)].head()
```

```python
# выведем min и max
ts_of_users = ratings.groupby('userId').agg({'timestamp': ['min', 'max']}).reset_index()
ts_of_users.head()
```

```python
def avg_ts_of_user(row):
    """функция выводит разницу между min и max timestamp в сутках.
    Предполагается, что timestamp всегда больше или равно 0"""
    n = row['timestamp']['max'] - row['timestamp']['min']
    return n/60/60/24
```

```python
#применим функцию
ts_of_users['avg_timestamp_in_days'] = ts_of_users.apply(avg_ts_of_user, axis=1)
ts_of_users.head()
```

```python
#проверим максимальное значение на адекватность:
ts_of_users['avg_timestamp_in_days'].max()
```

```python
# 5969 дней - это сколько лет?
(ts_of_users['avg_timestamp_in_days'].max())/365
```

```python
#напишем функцию
```

```python
def production_year(row):
    """каждой строке из названия фильма выставляет год выпуска"""
    for row in row['title']:
        for n in year:
            if str(n) in row:
                return n
            else:
                return '1900'
```
