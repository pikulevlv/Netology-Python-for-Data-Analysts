4. Визуализировать распределение переменной ‘Price’, рассчитать показатели skewness и kurtosis

```python
plt.hist(data['price'], bins=50)
plt.figure(figsize=(15, 8))
plt.ylabel('распределение показателя price')
plt.xlabel('$price$')
```

```python
sts.kurtosis(data['price'])
```

```python
sts.skew(data['price'])
```

```python
print( 'Эксцесс: {}'.format( sts.kurtosis(data['price']) ))
print( 'Асимметрия : {}'.format( sts.skew(data['price']) ) )
```
