#### Посчитать руками статистики $\bar{X}, s^2, \mu, MODE, X_{0.75} - X_{0.25}$ для поля mean symmetry

```python
# визуально видно, что полей со словом 'symmetry' в названии несколько, найдем их индексы
columns = list(df.columns)
for column in columns:
    if 'symmetry' in column:
        print(column, columns.index(column))
```

```python
# your solution is here
# считаем статистики руками
ms = df['mean symmetry']
ms_mean = ms.mean()
ms_s2 = sum([(ms_elem - ms_mean)**2 for ms_elem in ms]) / (len(ms) - 1)
ms_s = (sum([(ms_elem - ms_mean)**2 for ms_elem in ms]) / (len(ms)-1)) ** 0.5
ms_mode = ms.max()
ms075_025 = list(set(sorted(ms)[: int(len(ms)*0.75 + 1)]) - set(sorted(ms)[: int(len(ms)*0.25 + 1)]))
ms_mu = (float(sorted(ms)[len(ms)//2 - 1]) + float(sorted(ms)[len(ms)//2])) / 2
X75 = sorted(ms)[: int(len(ms)*0.75 + 1)][-1]
X25 = sorted(ms)[: int(len(ms)*0.25 + 1)][-1]
```

```python
print(f"mean (выборочное среднее): {ms_mean}")
print(f"s2 (выборочная дисперсия) : {ms_s2}")
print(f"mode (мода): {ms_mode}")
print(f"X 75 (75%-й квантиль) : {X75}")
print(f"X 25 (25%-й квантиль) : {X25}")
print(f"mu (медиана, 50%-й квантиль) : {ms_mu}")
```
